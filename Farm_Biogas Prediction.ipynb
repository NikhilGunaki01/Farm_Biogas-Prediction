# <div style="padding: 30px; color:white; margin:10; font-size:150%; text-align:left; display:fill; border-radius:10px; background-color:#3b3745"><b><span style='color:#F1A424'>3 |</span></b> <b>Data Cleaning</b></div>

```python
drop_columns = ["Project Name", "City", "County", "State", 'Cattle', 'Poultry', 'Swine', 'Total_Animals']

df.drop(drop_columns, axis=1, inplace=True)

df = df[df['Dairy'] != 0]  # Focus only on dairy cows
```

```python
df.head()
```

```python
df.shape
```

```python
# Identify Numerical and Categorical Features
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
num_cols.remove("Biogas Generation Estimate (cu-ft/day)")
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
# Encoding Categorical Variables

df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
# Correlation Analysis

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

# <div style="padding: 30px; color:white; margin:10; font-size:150%; text-align:left; display:fill; border-radius:10px; background-color:#3b3745"><b><span style='color:#F1A424'>4 |</span></b> <b>Data Preprocessing</b></div>

```python
# Scaling Numerical Features

scaler = RobustScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# Split Data

y = df["Biogas Generation Estimate (cu-ft/day)"]
X = df.drop(["Biogas Generation Estimate (cu-ft/day)"], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=17)
```

# <div style="padding: 30px; color:white; margin:10; font-size:150%; text-align:left; display:fill; border-radius:10px; background-color:#3b3745"><b><span style='color:#F1A424'>5 |</span></b> <b>Model Development</b></div>

```python
# Model List

models = [('LR', LinearRegression()),
          ("Ridge", Ridge()),
          ("Lasso", Lasso()),
          ("ElasticNet", ElasticNet()),
          ('KNN', KNeighborsRegressor()),
          ('CART', DecisionTreeRegressor()),
          ('RF', RandomForestRegressor()),
          ('GBM', GradientBoostingRegressor()),
          ("XGBoost", XGBRegressor(objective='reg:squarederror')),
          ("LightGBM", LGBMRegressor()),
          ("CatBoost", CatBoostRegressor(verbose=False))]
```

```python
# Train and Evaluate Models

rmse_scores = []
r2_scores = []
mae_scores = []
mse_scores = []
execution_times = []
model_names = []
```

```python
for name, regressor in models:
    start_time = time.time()
    regressor.fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    execution_time = time.time() - start_time
    
    rmse_scores.append(rmse)
    r2_scores.append(r2)
    mae_scores.append(mae)
    mse_scores.append(mse)
    execution_times.append(execution_time)
    model_names.append(name)
```

```python
results_df = pd.DataFrame({
    'Model': model_names,
    'RMSE': rmse_scores,
    'R^2 Score': r2_scores,
    'MAE': mae_scores,
    'MSE': mse_scores,
    'Execution Time (s)': execution_times
})

results_df
```

# <div style="padding: 30px; color:white; margin:10; font-size:150%; text-align:left; display:fill; border-radius:10px; background-color:#3b3745"><b><span style='color:#F1A424'>6 |</span></b> <b>Hyperparameter Optimization</b></div>

```python
param_grids = {
    'RF': {'n_estimators': [10, 30, 50, 70, 100], 'max_depth': [None, 5, 10, 20]},
}

best_model = None
best_score = np.inf
```

```python
for name, regressor in models:
    if name in param_grids:
        grid_search = GridSearchCV(regressor, param_grid=param_grids[name], cv=5, n_jobs=-1)
        grid_search.fit(X_train, y_train)
        best_regressor = grid_search.best_estimator_
    else:
        best_regressor = regressor
        best_regressor.fit(X_train, y_train)
    
    y_pred = best_regressor.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    if rmse < best_score:
        best_score = rmse
        best_model = best_regressor
```

```python
print("Best Model:", best_model)
```
